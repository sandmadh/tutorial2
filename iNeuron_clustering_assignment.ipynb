```python
import numpy as np
from sklearn import cluster, datasets
from scipy import misc
import scipy as sp
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
face = sp.misc.face(gray = True)
```

```python
face.shape
```

```python
plt.figure(figsize=(10, 4))
plt.imshow(face, cmap=plt.cm.gray)
plt.show()
```

```python
new_face = face.reshape(-1,1)
```

```python
new_face.shape
```

```python
n_clusters = 5
kmeans = KMeans(n_clusters=5,random_state =42)
```

```python
kmeans.fit(new_face)
```

```python
#to get the new imaege
labels = kmeans.labels_
```

```python
type(labels)
```

```python
labels.shape
```

```python
compressed_image = labels.reshape(face.shape[0],face.shape[1])
```

```python
compressed_image.shape
```

```python
# lets plot the compressed image
plt.figure(figsize=(10, 4))
plt.imshow(compressed_image, cmap=plt.cm.gray)
plt.show()
```
